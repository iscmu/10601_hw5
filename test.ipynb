### Task 3.1

```python
import numpy as np
from neuralnet import NN, zero_init, random_init
import matplotlib.pyplot as plt
train_input = "small_train.csv"
X_tr = np.loadtxt(train_input, delimiter=',')
y_tr = X_tr[:, 0].astype(int)
X_tr = X_tr[:, 1:]  # cut off label column

validation_input = "small_validation.csv"
X_te = np.loadtxt(validation_input, delimiter=',')
y_te = X_te[:, 0].astype(int)
X_te = X_te[:, 1:]  # cut off label column

labels = ["a", "e", "g", "i", "l", "n", "o", "r", "t", "u"]

n_hidden = [5, 20, 50, 100, 200]
lr = 0.001
n_epochs = 100
init_flag = 1
```

```python
losses_tr = []
losses_te = []
for n_hid in n_hidden:
    nn = NN(
        input_size=X_tr.shape[-1],
        hidden_size=n_hid,
        output_size=len(labels),
        weight_init_fn=zero_init if init_flag == 2 else random_init,
        learning_rate=lr
    )
    train_losses, test_losses = nn.train(X_tr, y_tr, X_te, y_te, n_epochs)
    test_labels, test_error_rate = nn.test(X_te, y_te)
    losses_tr.append(train_losses[-1])
    losses_te.append(test_losses[-1])
```

```python
fig, ax = plt.subplots()
# 4. computer generated line graphs/curves
ax.plot(n_hidden, losses_tr, label="training")
ax.plot(n_hidden, losses_te, label='validation')
# 3. provide a legend
ax.legend()
# 1. title your graph
ax.set_title("Number of hidden units against average loss over 100 epochs")
# 2. label your axes
ax.set_xlabel("Number of hidden units")
ax.set_ylabel("Average cross-entropy loss")
plt.show()
```

### Task 3.2

```python
import numpy as np
from neuralnet import NN, zero_init, random_init
import matplotlib.pyplot as plt
train_input = "small_train.csv"
X_tr = np.loadtxt(train_input, delimiter=',')
y_tr = X_tr[:, 0].astype(int)
X_tr = X_tr[:, 1:]  # cut off label column

validation_input = "small_validation.csv"
X_te = np.loadtxt(validation_input, delimiter=',')
y_te = X_te[:, 0].astype(int)
X_te = X_te[:, 1:]  # cut off label column

labels = ["a", "e", "g", "i", "l", "n", "o", "r", "t", "u"]

n_hidden = [5, 20, 50, 100, 200]
lr = [0.05, 0.005, 0.0005]
n_epochs = 100
init_flag = 1

def plot_curve(lr):
    losses_tr = []
    losses_te = []
    for n_hid in n_hidden:
        nn = NN(
            input_size=X_tr.shape[-1],
            hidden_size=n_hid,
            output_size=len(labels),
            weight_init_fn=zero_init if init_flag == 2 else random_init,
            learning_rate=lr
        )
        train_losses, test_losses = nn.train(X_tr, y_tr, X_te, y_te, n_epochs)
        test_labels, test_error_rate = nn.test(X_te, y_te)
        losses_tr.append(train_losses[-1])
        losses_te.append(test_losses[-1])
    fig, ax = plt.subplots()
    ax.plot(n_hidden, losses_tr, label="training")
    ax.plot(n_hidden, losses_te, label='validation')
    ax.legend()
    ax.set_title(f"Number of hidden units against average loss over 100 epochs, lr={lr}")
    ax.set_xlabel("Number of hidden units")
    ax.set_ylabel("Average cross-entropy loss")

for i in lr:
    plot_curve(i)
```

```python
import numpy as np
train_input = "small_train.csv"
X_tr = np.loadtxt(train_input, delimiter=',')
y_tr = X_tr[:, 0].astype(int)
X_tr = X_tr[:, 1:]  # cut off label column

validation_input = "small_validation.csv"
X_te = np.loadtxt(validation_input, delimiter=',')
y_te = X_te[:, 0].astype(int)
X_te = X_te[:, 1:]  # cut off label column

labels = ["a", "e", "g", "i", "l", "n", "o", "r", "t", "u"]

n_hid = 50
lr = 0.003
n_epochs = 100
init_flag = 1
```

```python
from neuralnet import NN, zero_init, random_init
from neuralnett34 import NN2

nn = NN(
    input_size=X_tr.shape[-1],
    hidden_size=n_hid,
    output_size=len(labels),
    weight_init_fn=zero_init if init_flag == 2 else random_init,
    learning_rate=lr
)

nn2 = NN2(
    input_size=X_tr.shape[-1],
    hidden_size=n_hid,
    output_size=len(labels),
    weight_init_fn=zero_init if init_flag == 2 else random_init,
    learning_rate=lr
)

train_losses1, test_losses1 = nn.train(X_tr, y_tr, X_te, y_te, n_epochs)
train_losses2, test_losses2 = nn2.train(X_tr, y_tr, X_te, y_te, n_epochs)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
# 4. computer generated line graphs/curves
ax.plot(np.arange(n_epochs), train_losses1, label="1-hidden: training")
ax.plot(np.arange(n_epochs), test_losses1, label="1-hidden: validation")
ax.plot(np.arange(n_epochs), train_losses2, label="2-hidden: training")
ax.plot(np.arange(n_epochs), test_losses2, label="2-hidden: validation")
# 3. provide a legend
ax.legend()
# 1. title your graph
ax.set_title("Comparison of 1-hidden and 2-hidden NN")
# 2. label your axes
ax.set_xlabel("Epoch number")
ax.set_ylabel("Average cross-entropy loss")
plt.show()
```
